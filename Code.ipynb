```python
import csv

def writeComment(name, grade, pl, tl, al, period, op):
  vowel = 'a e i o u'
  vowels = vowel.split()
  a1 = 'a'
  a2 = 'a'
  a3 = 'a'
  if pl[0] in vowels:
    a1 = 'an'
  if tl[0] in vowels:
    a2 = 'an'
  if grade[0] in vowels:
    a3 = 'an'
  closer = 'keep up the great work'
  if tl in ['low', 'passable']:
    closer = 'theres always room for improvement'
  comment = (f'{op}\n\n{name}, you have {a1} {pl} level of engagement and competency in this class during this {period}. You turn in homework at {a2} {tl} rate, {closer}! You are {al} at being on time and engaged to class, keep it up, but a little improvement is never a bad idea.\n\n{name}, given your performance this {period}, you are receiving {a3} {grade} in this class, I hope that you believe this reflects your experience during this {period}.\n\n---\n\n')
  return comment

with open('comments.txt', 'w') as com:
  with open('teacher.csv') as csvfile:
    with open('op.txt') as oppener:
      op = str(oppener.readlines()[0])
    INFO = csv.reader(csvfile, delimiter=',')
    HEADER = next(INFO)
    for row in INFO:
      TL = ['low', 'passable', 'decent', 'high', 'perfect']
      PL = ['relatively poor', 'acceptable', 'decent', 'high', 'exceptional']
      AL = ['lousy', 'okay', 'on par', 'above average', 'exceptional']
      name = row[0]
      grade = row[-1]
      period = row[2]
      pl = PL[int(row[1])-1]
      tl = TL[int(row[4])-1]
      al = AL[int(row[5])-1]
      pronoun = row[3]
      y = str(writeComment(name, grade, pl, tl, al, period, op))
      print(y)
      com.write(y)
```
